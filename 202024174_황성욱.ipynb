```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
```

```python
iris = pd.read_csv('iris.csv')
iris.info()
```

```python
iris.describe()
```

```python
iris['species'].value_counts()
```

```python
iris.isna().sum()
```

```python
iris.duplicated().sum()
```

```python
iris[iris.duplicated()]
```

```python
iris[iris.duplicated(keep=False)]
```

```python
iris = iris.drop_duplicates(keep='first')

```

```python
iris[iris.duplicated()]
```

```python
iris.groupby('species').sum()
```

```python
avg_sepal_length = iris.groupby('species').mean()
avg_sepal_length
```

```python
plt.rc('font', family='Gulim')
sample = iris['sepal_length'][:10]
indices = range(len(sample))
plt.bar(indices, sample)

plt.title('꽃받침 길이 시각화')
plt.xlabel('Data')
plt.ylabel('꽃받침의 길이')
plt.show()
```

```python
plt.rc('font', family='Gulim')
sns.barplot(x='species', y='sepal_length', data= iris)
plt.xlabel('종류')
plt.ylabel('꽃받침 길이의 평균값')
plt.title('품종별 sepal length 그래프')
plt.show()
```

```python
plt.rc('font', family='Gulim')
plt.hist(iris['sepal_length'], bins=3, color='blue', align='mid', label='Sepal Length')
plt.hist(iris['sepal_width'], bins=3, color='orange', align='mid', label='Sepal Width')
plt.hist(iris['petal_length'], bins=3, color='green', align='mid', label='Petal Length')
plt.hist(iris['petal_width'], bins=3, color='red', align='mid', label='Petal Width')



plt.title('각 속성 값들의 빈도수 히스토그램')
plt.xlabel('Value')
plt.ylabel('빈도수')
plt.legend()
plt.show()
```

```python
plt.scatter(iris['sepal_length'], iris['petal_length'])
plt.xlabel('sepal_length')
plt.ylabel('petal_length')
plt.show()
```

```python
titanic = pd.read_csv('titanic.csv')
titanic
```

```python
titanic.info()
```

```python
titanic.describe()
```

```python
titanic.sort_values(by=['fare'], ascending=False)
```

```python
titanic['survived'].value_counts()
```

```python
titanic.isna().sum()
```

```python
titanic.drop('deck', axis=1, inplace=True)
print(titanic)
```

```python
titanic['embarked'].value_counts()
```

```python
titanic.fillna({'embarked':'S'}, inplace=True)
titanic.isna().sum()
```

```python
avg_age = titanic['age'].mean()
titanic.fillna({'age':avg_age}, inplace=True)
titanic.isna().sum()
```

```python
titanic.groupby('embarked')['pclass'].mean()
```

```python
titanic.groupby(['pclass', 'sex']).mean('age')
```

```python
sns.countplot(x='pclass', hue=titanic['survived'], data=titanic)
plt.show()
```

```python
numeric_cols = titanic.select_dtypes(include=[np.number])
correlation = numeric_cols.corr()
sns.heatmap(data=correlation, annot=True, fmt='.2f', cbar=False)
plt.show()
```
