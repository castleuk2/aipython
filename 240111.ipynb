```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
fig, axs = plt.subplots(2,2)
axs[0,0].plot(x,y)
axs[0,1].scatter(x,y)
axs[1,0].bar(x,y)
axs[1,1].barh(x,y)
fig.suptitle('Drawing Graph')
plt.show()
```

```python
import seaborn as sns
```

```python
iris = sns.load_dataset("iris")
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=iris)
plt.title('Scatter Plot by seaborn', fontsize = 20)
plt.show()
```

```python
iris_data = sns.load_dataset('iris')
sns.countplot(x='species', data=iris_data)
plt.show()
```

```python
sns.barplot(x = 'species', y = 'petal_length', data = iris_data, errorbar='sd')
plt.show()
```

```python
sns.pairplot(iris_data, hue='species')
plt.show()
```

```python
sns.regplot(x='petal_length',y='petal_width',data=iris_data)
plt.show()
```

```python
#MISSON 3
penguins_data = sns.load_dataset('penguins')
sns.barplot(x='species', y='body_mass_g', data=penguins_data, hue='sex')
plt.show()
```

```python
sns.pairplot(penguins_data, hue='species')
plt.show()
```

```python
sns.regplot(x='body_mass_g', y='flipper_length_mm', data= penguins_data)
plt.show()
```
